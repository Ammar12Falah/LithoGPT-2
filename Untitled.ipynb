```python
import io, zipfile
from pathlib import Path

def unpack_kgs_las(raw_root="data/raw"):
    arch_dir = Path(raw_root) / "kgs" / "archives"
    out = Path(raw_root) / "kgs" / "las"
    out.mkdir(parents=True, exist_ok=True)
    existing = {p.name for p in out.glob("*.las")}
    n_new = 0

    def _junk(name):
        b = Path(name).name
        return b.startswith("._") or "__MACOSX" in name

    def _write(name, data):
        nonlocal n_new
        (out / name).write_bytes(data)
        existing.add(name)
        n_new += 1

    for zpath in sorted(arch_dir.glob("*.zip")):
        with zipfile.ZipFile(zpath) as zf:
            for member in zf.namelist():
                if _junk(member):
                    continue
                low = member.lower()
                if low.endswith(".las"):                        # 2024-style: direct .las
                    name = Path(member).name
                    if name not in existing:
                        _write(name, zf.read(member))
                elif low.endswith(".zip"):                       # 2014-style: nested per-well zip
                    stem = Path(member).stem
                    if f"{stem}.las" in existing:
                        continue
                    try:
                        inner = zipfile.ZipFile(io.BytesIO(zf.read(member)))
                    except zipfile.BadZipFile:
                        continue
                    las = [m for m in inner.namelist() if not _junk(m) and m.lower().endswith(".las")]
                    for k, m in enumerate(las):
                        name = f"{stem}.las" if k == 0 else f"{stem}_{k}.las"
                        if name not in existing:
                            _write(name, inner.read(m))
    print(f"[kgs] unpacked {n_new} new LAS files to {out}")
    return n_new

unpack_kgs_las()
```

```python
import lasio, numpy as np, glob, random
targets = ["POR(NEU)","CNLS","RHOC","RLL3","DEN(CDL)","GAMMA"]
vals = {t: [] for t in targets}
files = glob.glob("data/raw/kgs/las/*.las"); random.shuffle(files)
for f in files[:400]:
    try: las = lasio.read(f)
    except Exception: continue
    for c in las.curves:
        if c.mnemonic in targets:
            d = np.asarray(c.data, float); d = d[np.isfinite(d)]
            if d.size: vals[c.mnemonic].extend(list(d[np.abs(d)<1e5])[:200])
for t in targets:
    a = np.array(vals[t])
    print(f"{t:10} n={a.size:6}", "" if a.size==0 else
          f"p5={np.percentile(a,5):.2f} med={np.percentile(a,50):.2f} p95={np.percentile(a,95):.2f}")
```

```python
from pathlib import Path

# 1. correct the two aliases the audit caught
p = "configs/mnemonic_aliases.yaml"; s = open(p).read()
s = s.replace(', "RHOC"', '')                      # density-correction curve, not bulk density
s = s.replace(', "RLL3"]', ']')                    # remove RLL3 from RDEP
s = s.replace('"RES(MI)"]', '"RES(MI)", "RLL3"]')  # Laterolog-3 is moderate depth, belongs in RMED
open(p, "w").write(s)

# 2. write the audit record the gate needs
Path("reports/alias_audit").mkdir(parents=True, exist_ok=True)
open("reports/alias_audit/AUDIT.md", "w").write("""# KGS alias unit and physics audit

Value ranges from a 400-well sample of the on-disk 2016 KGS LAS, 5 July 2026.

- DEN(CDL): median 2.44 g/cc, bulk density, mapped to RHOB, kept.
- GAMMA: median 20 API, gamma ray, mapped to GR, kept.
- RHOC: median 0.06, range about -0.5 to 0.3. This is a density-correction curve,
  not bulk density. Removed from RHOB.
- RLL3: median 4.37 ohm-m, resistivity. A Laterolog-3 is a moderate-investigation
  device, so it was moved from RDEP to RMED.
- CNLS: median 35 (percent), neutron porosity, mapped to NPHI. Curves with a
  percent unit convert to fraction correctly; curves with a blank unit do not
  convert and gate out. A percent-default rule for KGS neutron and the
  limestone-matrix assumption are recorded as G2 harmonization tasks.
- POR(NEU): absent from the 2016 sample, present in the deleted 2024 and 2014
  years, so its audit is deferred to the G2 reprocess.

Count impact: none on the 5,644 G1 total. RHOC values fall below the RHOB range
and contributed no coverage. Moving RLL3 relabels a resistivity curve rather than
removing one. The corrected config takes effect at the G2 reprocess.
""")

# 3. record the gate-bar provenance in the G1 report
with open("reports/status_g1.md", "a") as f:
    f.write("\n## Gate bar provenance\n"
            "The 5,000 QC-passing bar is set by docs/HANDOFF.md Section 11 timeline, "
            "week 2 row: 5,000 floor, 8,000 stretch target. Both predate execution. "
            "Design authority conditional approval recorded 5 July 2026.\n")
print("config corrected, audit written, provenance recorded")
```

```python
import re, ast
p = "src/lithogpt2/ingest/kgs.py"; s = open(p).read()
new = '''def unpack_las(raw_root: str = "data/raw") -> int:
    # Extract .las from archive ZIPs into kgs/las/. Handles direct .las and
    # nested zip-of-zips (one inner .zip per well); skips macOS junk; idempotent.
    dest = Path(raw_root) / SOURCE
    arch_dir = dest / "archives"
    out = dest / "las"
    out.mkdir(parents=True, exist_ok=True)
    existing = {q.name for q in out.glob("*.las")}
    n = 0
    def _junk(name):
        return Path(name).name.startswith("._") or "__MACOSX" in name
    def _write(name, data):
        nonlocal n
        (out / name).write_bytes(data); existing.add(name); n += 1
    for zpath in sorted(arch_dir.glob("*.zip")):
        with zipfile.ZipFile(zpath) as zf:
            for member in zf.namelist():
                if _junk(member):
                    continue
                low = member.lower()
                if low.endswith(".las"):
                    name = Path(member).name
                    if name not in existing:
                        _write(name, zf.read(member))
                elif low.endswith(".zip"):
                    stem = Path(member).stem
                    if f"{stem}.las" in existing:
                        continue
                    try:
                        inner = zipfile.ZipFile(io.BytesIO(zf.read(member)))
                    except zipfile.BadZipFile:
                        continue
                    las = [m for m in inner.namelist() if not _junk(m) and m.lower().endswith(".las")]
                    for k, m in enumerate(las):
                        name = f"{stem}.las" if k == 0 else f"{stem}_{k}.las"
                        if name not in existing:
                            _write(name, inner.read(m))
    print(f"[kgs] unpacked {n} LAS files to {out}")
    return n
'''
s = re.sub(r"def unpack_las\(.*?\n    return n\n", new, s, count=1, flags=re.S)
open(p, "w").write(s); ast.parse(open(p).read())
print("nested unpacker in kgs.py:", "nested zip-of-zips" in open(p).read())
```

```python
import subprocess, shutil, os, glob
from pathlib import Path

# protect the committed G1 evidence CSVs
saved = {}
for f in ["kgs_qc_records.csv","kgs_coverage.csv","kgs_unmapped_mnemonics.csv","kgs_failures.csv"]:
    src = Path("reports")/f
    if src.exists():
        saved[f] = f"/tmp/save_{f}"; shutil.copy(src, saved[f])

shutil.rmtree("data/processed/kgs", ignore_errors=True)

# reprocess 500 on-disk 2016 wells WITH parquets on (no --no-parquet)
r = subprocess.run(["python","scripts/run_qc_kgs.py","--fresh","--max-wells","500"],
                   capture_output=True, text=True)
print(r.stdout[-250:]); print("ERR:", r.stderr[-200:])

pq = glob.glob("data/processed/kgs/*.parquet")
total = sum(os.path.getsize(x) for x in pq); n = len(pq)
per = total/n if n else 0
print(f"\nparquets written: {n}  total {total/1e6:.1f} MB  per-well {per/1e6:.3f} MB")
print(f"projected for full corpus (~9,600 wells KGS+NLOG): {per*9600/1e9:.2f} GB")
print(os.popen("df -h /workspace | tail -1").read())

# restore G1 evidence; remove the sample parquets (regenerable)
for f, bak in saved.items():
    shutil.copy(bak, Path("reports")/f)
shutil.rmtree("data/processed/kgs", ignore_errors=True)
print("G1 records restored, sample parquets removed")
```
